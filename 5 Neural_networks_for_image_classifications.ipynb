### Building and training the neural network

```
(16384 + 2) / 2
```

```
# 16384 -> 8193 -> 8193 -> 1
network1 = tf.keras.models.Sequential()
network1.add(tf.keras.layers.Dense(input_shape=(16384,), units=8193, activation='relu'))
network1.add(tf.keras.layers.Dense(units=8193, activation = 'relu'))
network1.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```
# https://keras.io/api/optimizers/
# https://keras.io/api/losses/
network1.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```
history = network1.fit(X_train, y_train, epochs=30)
```

## Approach 2 - feature extraction

### Feature extractor

```
files = [os.path.join(directory, f) for f in sorted(os.listdir(directory))]
print(files)
```

```
export = 'mouth,lips,class\n'
```

```
show_images = False
features = []
```

```
100 * 200
```

```
(2000 / 20000) * 100
```

```
100 / 2
```

```
for image_path in files:
  #print(image_path)
  try:
    original_image = cv2.imread(image_path)
    (H, W) = original_image.shape[:2]
  except:
    continue

  image = original_image.copy()
  image_features = []
  mouth =  0
  lips = 0

  image_name = os.path.basename(os.path.normpath(image_path))

  if image_name.startswith('N'):
    class_name = 0
  else:
    class_name = 1

  for height in range(0, H):
    for width in range(0, W):
      blue = image.item(height, width, 0)
      green = image.item(height, width, 1)
      red = image.item(height, width, 2)

      # Tony- black beard
      if height > (H / 2):
       if (blue >= 0 and blue <= 20 and green >= 0 and green <= 20 and red >=0 and red <= 20):
        image[height, width] = [0, 255, 255]
        mouth += 1

      

      #  Natasha- red lips
      if height > (H / 2):
       if (blue >= 30 and blue <= 140 and green >= 20 and green <= 100 and red >= 220 and red <= 255):
        image[height, width] = [0, 255, 128]
        lips += 1

      

  mouth = round((mouth / (H * W)) * 100, 9)
  

  lips = round((lips / (H * W)) * 100, 9)
  
  image_features.append(mouth)
  
  image_features.append(lips)
  
  image_features.append(class_name)

  features.append(image_features)

 
  f = (",".join([str(item) for item in image_features]))
  export += f + '\n'

  if show_images == True:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    fig, im = plt.subplots(1, 2)
    im[0].axis('off')
    im[0].imshow(original_image)
    im[1].axis('off')
    im[1].imshow(image)
    plt.show()
```

```
export
```

```
with open('features.csv', 'w') as file:
  for l in export:
    file.write(l)
file.closed
```
